```python
import pandas as pd
import numpy as np
from ase import Atoms
from ase.visualize import view
import ast

# CSV 파일 읽기
df = pd.read_csv('/home/minkyu/DiffCSP/data/oc20_2M/test.csv')

id = 30

data = df.iloc[id]

atom_types = ast.literal_eval(data['atom_types'])
lengths = ast.literal_eval(data['lengths'])
angles = ast.literal_eval(data['angles'])
frac_coords = ast.literal_eval(data['frac_coords'])
cell = lengths + angles

atoms = Atoms(symbols=atom_types, scaled_positions=frac_coords, cell=cell, pbc=True)
 
print(data['num_atoms'])
view(atoms, viewer='ngl')  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')  # seaborn의 정확한 스타일 이름 사용

# CSV 파일 읽기
df = pd.read_csv('data/oc20_2M/train.csv')

# 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(df['slab_num_atoms'], bins=50, edgecolor='black')
plt.title('Distribution of Number of Atoms', fontsize=14)
plt.xlabel('Number of Atoms', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)

# 그래프 표시
plt.show()
```
